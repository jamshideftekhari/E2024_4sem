```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

data forberedelse. .reshape(-1, 1), fordi scikit-learn forventer, at dataene har en bestemt form. Her sikrer vi, at vægten er i en kolonne.

```python
# Data: vægt i kg og pris i kr.
vaegt = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Vægt i kg
print(vaegt)
pris = np.array([150, 300, 450, 600, 750])        # Pris i kr.
print(pris) 
```

Nu vil vi oprette en lineær regressionsmodel og træne den med vores data.

```python
# Opret modellen
model = LinearRegression()

# Træn modellen med dataene
model.fit(vaegt, pris)
```

Lad os nu bruge modellen til at forudsige prisen på en vare, der vejer 6 kg.

```python
# Forudsige prisen for en vare på 6 kg
vaegt_ny = np.array([[6]])  # Den nye vægt vi vil forudsige
pris_forudsigelse = model.predict(vaegt_ny)

print(f"Den forudsagte pris for en vare på 6 kg er: {pris_forudsigelse[0]:.2f} kr.")
```

visualize the data and the model's prediction line.

```python
import matplotlib.pyplot as plt

# Plot the data points
plt.scatter(vaegt, pris, color="blue", label="Data points")

# Plot the regression line
plt.plot(vaegt, model.predict(vaegt), color="red", label="Regression line")

# Plot the predicted point for 6 kg
plt.scatter(6, pris_forudsigelse, color="green", label="Prediction for 6 kg")

# Labels and legend
plt.xlabel("Vægt (kg)")
plt.ylabel("Pris (kr)")
plt.legend()
plt.show()
```
